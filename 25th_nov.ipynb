```python
#prime number
#Write a program to display the numbers sequentially from 1 to 99 with 5 numbers on  each line  

i = 1
while i <= 99:
    for j in range(5):   # print 5 numbers per line
        if i <= 99:
            print(i, end=" ")
            i += 1       # increment i each time
    print()              # move to next line after 5 numbers
```

```python
# res =ft(row)/(ft(row-col_two))*ft(col_two)
#print(int(res), end= "")n = int(input())
from math import factorial as ft

n = int(input("enter a value: "))
for i in range(n):
    for j in range(n - i-1):
        print(' ' , end = " ")
    for k in range(i+1):
        res =ft(i)/(ft(i-k) * ft(k))
        print(int(res), end= " ")
    print()
```
