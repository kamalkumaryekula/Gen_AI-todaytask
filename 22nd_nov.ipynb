```python
'''Write a program to calculate the monthly income of a person using the following 
commission schedule : (use ‘if-else-if’ statement).  '''

sales =  int(input("enter monthly sales:"))

if sales >= 50000:
    commission = (sales*0.16) + 375
elif 50000 <= sales >= 40000:
    commission = (sales*0.14) + 350
elif 40000 <= sales >= 30000:
    commission = (sales*0.12) + 325
elif 30000 <= sales >= 20000:
    commission = (sales*0.09) + 300
elif 20000 <= sales >= 10000:
    commission = (sales*0.05) + 250
else:
    commission = (sales*0.03) + 200

print(f"commisssion for {sales} sale is: {commission}")
```

```python
# Program to check if the middle digit of a 3-digit number
# is equal to the sum of the other two digits

num = int(input("Enter a 3-digit number: "))

# Convert to string to easily access digits
num_str = str(num)

if len(num_str) == 3:
    print("It is a three digit number")
    first = int(num_str[0])
    middle = int(num_str[1])
    last = int(num_str[2])

    if middle == first + last:
        print("True: Middle digit equals sum of other two digits")
    else:
        print("False: Middle digit does not equal sum of other two digits")
else:
    print("It is not a three digit number")
```

```python
char = input("Enter a vowel character: ")

match char:
    case "a" | "A":
        print("Apple")
    case "e" | "E":
        print("Elephant")
    case "i" | "I":
        print("Ice cream")
    case "o" | "O":
        print("Orange")
    case "u" | "U":
        print("Umbrella")
    case _:
        print("Not a vowel character")
```

```python
num_1 = float(input("Enter first number: "))
num_2 = float(input("Enter second number: "))
operator = input("Enter an operator: ")

match operator:
    case "+":
        print("Result:", num_1 + num_2)
    case "-":
        print("Result:", num_1 - num_2)
    case "*":
        print("Result:", num_1 * num_2)
    case "/":
        if num_2 != 0:
            print("Result:", num_1 / num_2)
        else:
            print("Error: Division by zero")
    case _:
        print("Invalid operator")
```
