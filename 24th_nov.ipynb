```python
#Write a program to read a character and find out whether it is uppercase or lowercase
char = input("Enter a character: ")
if char.islower():
    print("it is a lower case letter")
elif char.isupper() :
    print("it is upper case letter")
else:
    print("it is not an alphabet")
```

```python
#Write a program to find the sum of ’n’ distinct numbers. 
n = int(input())
numbers = []
for i in range (n):
    num = int(input())
    if num in numbers:
        print("it is already existed")
    else:
        numbers.append(num)
print(sum(numbers))
    
```

```python
#Write a program to find the sum of even ‘n’ natural numbers. 
sum = 0
#count = 0
n = int(input())
for i in range(n):
    if i%2 ==0:
        sum += i
      #  count +=1
print(f"sum is: {sum} ")    
```
